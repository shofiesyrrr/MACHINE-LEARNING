```python
# membaca file csv menggunakan pandas
import pandas as pd
from google.colab import drive

# Remount Google Drive
drive.mount('/content/gdrive', force_remount=True)

df = pd.read_csv(path + 'data/day.csv')
df.head()
```

```python
# Hapus kolom yang tidak relevan/redundant
df = df.drop(['instant', 'dteday', 'casual', 'registered'], axis=1)

# 2. Praproses Data: Feature Engineering/Encoding (Multiple Linear Regression memerlukan)
# Konversi variabel kategori menjadi One-Hot Encoding
df = pd.get_dummies(df, columns=['season', 'yr', 'mnth', 'weekday', 'weathersit'], drop_first=True).astype(int)
# drop_first=True menghindari masalah multikolinearitas (Dummy Variable Trap)
```

```python
# Y: Variabel Dependen
Y = df['cnt']

# X: Variabel Independen (semua kolom, kecuali 'cnt')
X = df.drop('cnt', axis=1)

# PENTING: Untuk menghindari multikolinearitas antara temp dan atemp (keduanya suhu)
# Disarankan untuk memilih salah satu. Kita pertahankan 'temp' dan drop 'atemp'
X = X.drop('atemp', axis=1)
```

```python
from sklearn.model_selection import train_test_split

# Pembagian data 80% train, 20% test
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y,
    test_size=0.2,
    random_state=42 # Agar hasil konsisten seperti pada modul [cite: 232, 588]
)

print(f"Jumlah data train: {len(X_train)}")
print(f"Jumlah data test: {len(X_test)}")
```

```python
# Periksa tipe data dari semua kolom di DataFrame X_train dan Y_train
print("Tipe data Y_train:")
print(Y_train.dtype)

print("\nTipe data X_train:")
print(X_train.dtypes)
```

```python
# Y_train: Pastikan target tidak ada nilai non-numerik
# errors='coerce' akan mengubah nilai non-numerik menjadi NaN
Y_train = pd.to_numeric(Y_train, errors='coerce')

# Hapus baris NaN yang mungkin muncul (jika ada data kotor)
Y_train = Y_train.dropna()
```

```python
# X_train: Konversi semua kolom menjadi float
for col in X_train.columns:
    X_train[col] = pd.to_numeric(X_train[col], errors='coerce')

# Hapus baris NaN yang mungkin muncul akibat konversi (jika ada data kotor)
X_train = X_train.dropna()

# Pastikan X_train dan Y_train memiliki index yang sama setelah dropna (Penting!)
common_index = X_train.index.intersection(Y_train.index)
X_train = X_train.loc[common_index]
Y_train = Y_train.loc[common_index]
```

```python
import statsmodels.api as sm

# Pastikan X_train memiliki tipe data numerik sebelum menambahkan konstanta
# Langkah get_dummies sebelumnya seharusnya menangani hal ini, tetapi sebagai langkah pencegahan
# kami secara eksplisit mengonversi kolom yang relevan atau seluruh DataFrame jika diperlukan.
# Mengasumsikan langkah get_dummies sekarang mengonversi dengan benar ke int:
X_train_const = sm.add_constant(X_train)

# Konversikan kolom boolean menjadi integer secara eksplisit sebelum memodelkan data
# Ini adalah langkah pencegahan jika astype(int) pada get_dummies tidak cukup
for col in X_train_const.columns:
    if X_train_const[col].dtype == 'bool':
        X_train_const[col] = X_train_const[col].astype(int)


# Buat dan latih model OLS
model_ols = sm.OLS(Y_train, X_train_const).fit()

# Cetak parameter dan persamaan regresi
print(model_ols.params)
print("\n--- Persamaan Regresi ---")
print(f"Y = {model_ols.params['const']:.2f} + ({model_ols.params['temp']:.2f}*temp) + ...")
```

```python
import numpy as np
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# Tambahkan konstanta ke data uji [cite: 712, 713]
X_test_const = sm.add_constant(X_test)

# Prediksi menggunakan data uji
Y_pred_test = model_ols.predict(X_test_const)

# Hitung Metrik Evaluasi
r2 = r2_score(Y_test, Y_pred_test)
mae = mean_absolute_error(Y_test, Y_pred_test)
mse = mean_squared_error(Y_test, Y_pred_test)
rmse = np.sqrt(mse)

print("\n--- Evaluasi Model dengan Data Uji ---")
print(f"R-squared (Test): {r2:.4f}")
print(f"MAE (kg): {mae:.2f}")
print(f"RMSE (kg): {rmse:.2f}")

# Tampilkan ringkasan lengkap model OLS [cite: 624]
# print(model_ols.summary())
```
