# **1. Membuat dataset (dari tabel Excel):**


```python
import pandas as pd
import numpy as np

# Data training dari Excel
data = {
    "No":    [1, 2, 3, 4, 5, 6, 7, 8],
    "suhu":  [10, 25, 15, 20, 18, 20, 22, 24],   # Temperatur Udara (°C)
    "angin": [0,  0,  5,  3,  7, 10,  5,  6],   # Kecepatan Angin (km/jam)
    "kelas": ["Dingin", "Panas", "Dingin", "Panas",
              "Dingin", "Dingin", "Panas", "Panas"]  # Klasifikasi/Persepsi Marry
}

df = pd.DataFrame(data)
df
```

# **2. Hitung selisih & jarak ke data testing (suhu=16, angin=3)**

```python
# Data testing
suhu_uji = 16
angin_uji = 3

# Hitung selisih (dikuadratkan) dan jarak Euclidean
df["selisih_suhu"]  = (df["suhu"]  - suhu_uji) ** 2
df["selisih_angin"] = (df["angin"] - angin_uji) ** 2
df["jarak"] = np.sqrt(df["selisih_suhu"] + df["selisih_angin"])

df
```

# **3. Urutkan berdasarkan jarak, lalu voting K=4 dan K=5**

```python
# Urutkan dari jarak yang paling kecil
df_sorted = df.sort_values("jarak")
df_sorted
```

```python
def voting(k):
    topk = df_sorted.head(k)
    return topk["kelas"].value_counts()

print("Voting K=4")
print(voting(4))   # hasil: 2 dingin, 2 panas (imbang)

print("\nVoting K=5")
print(voting(5))   # hasil: 3 dingin, 2 panas → kelas = Dingin
```

```python
# Keputusan akhir (pakai K=5 sesuai catatan di Excel)
hasil_kelas = voting(5).idxmax()
print("Data uji (suhu=16, angin=3) diklasifikasikan sebagai:", hasil_kelas)
```
