```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Path dataset (ubah sesuai lokasi kamu simpan)
path = "/content/gdrive/MyDrive/praktikum_ml/praktikum01/data/day.csv"

# Membaca dataset
df_day = pd.read_csv(path)
print("Jumlah total data:", len(df_day))
df_day.head(), "\n"

# 1. Split menjadi Training (80%) dan Testing (20%)
train_set, test_set = train_test_split(df_day, test_size=0.2, random_state=42)

# 2. Dari Training, ambil 10% untuk Validation
train_set, val_set = train_test_split(train_set, test_size=0.1, random_state=42)
```

```python
# Jumlah Data Training
print("Jumlah data Training:", len(train_set))

# Tampilkan 5 Baris Pertama dari Set Data Training
print("\n=== Data Training ===")
train_set.head()
```

```python
# Jumlah Data Validation
print("Jumlah data Validation:", len(val_set))

# Tampilkan 5 Baris Pertama dari Set Data Validation
print("\n=== Data Validation ===")
val_set.head()
```

```python
# Jumlah Data Testing
print("Jumlah data Testing:", len(test_set))

# Tampilkan 5 Baris Pertama dari Set Data Testing
print("\n=== Data Testing ===")
test_set.head()
```

```python
import os
from sklearn.linear_model import LinearRegression
import joblib

# Pastikan folder models/ ada
os.makedirs("/content/gdrive/MyDrive/praktikum_ml/praktikum02/models", exist_ok=True)

# Misal kita latih model Linear Regression
# (contoh: variabel prediktor 'temp' untuk memprediksi 'cnt')
X_train = train_set[['temp']]   # fitur
y_train = train_set['cnt']      # target

model = LinearRegression()
model.fit(X_train, y_train)

# Simpan model ke folder models/
model_path = "/content/gdrive/MyDrive/praktikum_ml/praktikum02/models/model_linreg.pkl"
joblib.dump(model, model_path)

print(f"Model berhasil disimpan di: {model_path}")
```
