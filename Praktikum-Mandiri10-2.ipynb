# **1. IMPORT LIBRARY**

```python
import pandas as pd
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
```

# **2. Masukkan Data ke DataFrame**

```python
data = {
    "NIM":   ["TI001","TI002","TI003","TI004","TI005",
              "TI006","TI007","TI008","TI009","TI010"],
    "HASIL_SEBENARNYA": ["Lulus","Lulus","Lulus","Lulus","Lulus",
                         "Tidak Lulus","Tidak Lulus","Tidak Lulus","Tidak Lulus","Tidak Lulus"],
    "HASIL_PREDIKSI":   ["Lulus","Lulus","Lulus","Tidak Lulus","Tidak Lulus",
                         "Lulus","Tidak Lulus","Tidak Lulus","Tidak Lulus","Tidak Lulus"]
}

df = pd.DataFrame(data)
df
```

# **3. Buat Confusion Matrix (dalam bentuk tabel)**

```python
# Tentukan urutan label (positif = "Lulus")
labels = ["Lulus", "Tidak Lulus"]

cm = confusion_matrix(df["HASIL_SEBENARNYA"], df["HASIL_PREDIKSI"], labels=labels)

cm_df = pd.DataFrame(
    cm,
    index=[f"Actual {l}" for l in labels],
    columns=[f"Pred {l}" for l in labels]
)

cm_df
```

# **4. Hitung Accuracy, Precision, dan Recall**

```python
# Accuracy
acc = accuracy_score(df["HASIL_SEBENARNYA"], df["HASIL_PREDIKSI"])

# Precision dan Recall untuk kelas positif "Lulus"
prec = precision_score(df["HASIL_SEBENARNYA"], df["HASIL_PREDIKSI"],
                       pos_label="Lulus")
rec = recall_score(df["HASIL_SEBENARNYA"], df["HASIL_PREDIKSI"],
                   pos_label="Lulus")

print("Accuracy :", acc)
print("Precision:", prec)
print("Recall   :", rec)

print("\nDalam persen:")
print("Accuracy :", acc * 100, "%")
print("Precision:", prec * 100, "%")
print("Recall   :", rec * 100, "%")
```
